### Hearst patterns can be used to find concept classes and instances of concept classes in (large) text corpora. The classic Hearst patterns are give in Table 1.



### Write a regular expression for the Hearst patterns displayed above and extract a list of words and their concept class from the infection Corpus from the Notebook. Count how many supporting examples for each case have been found.


```python
import codecs
import re
from collections import Counter
from pprint import pprint
import glob



count1 = 0     # initializing the variables to store the count for each pattern
count2 = 0
count3 = 0
count4 = 0
count5 = 0

result_1 = [] # initializing the arrays to store the phrases 
result_2 = []
result_3 = []
result_4 = []
result_5 = []

filelist = glob.glob("infect/*.txt") #reading the files

for f in filelist:
    file = codecs.open(f,'r','utf8') #taking each file
    
    for line in file:
        line = line.strip() #taking each line in the file
        
#Hearst patterns are expressed using regular expressions. Then the matching phrases are counted and stored separately
        
        resultlist1 = re.findall(r'\b((\w+) (such as) (\w+) (and|or \w+)?)\b',line)    #Hearst pattern 1    
        if(len(resultlist1)>0):
            count1 = count1 + len(resultlist1)
            for result in resultlist1:
                result_1.append(result[0])
            
        resultlist2 = re.findall(r'\b((\w+)(,?) (especially) (\w+) (and|or \w+)?)\b',line) #Hearst pattern 2     
        if(len(resultlist2)>0):
            count2 = count2 + len(resultlist2)
            for result in resultlist2:
                result_2.append(result[0])
            
        resultlist3 = re.findall(r'\b((\w+)(,?) (including) (\w+) (and|or \w+)?)\b',line)   #Hearst pattern 3 
        if(len(resultlist3)>0):
            count3 = count3 + len(resultlist3)
            for result in resultlist3:
                result_3.append(result[0])
                
        resultlist4 = re.findall(r'\b((\w+)(,?) (and other) (\w+))\b',line)        #Hearst pattern 4
        if(len(resultlist4)>0):
            count4 = count4 + len(resultlist4)
            for result in resultlist4:
                result_4.append(result[0])
                
        resultlist5 = re.findall(r'\b((\w+)(,?) (or other) (\w+))\b',line)        #Hearst pattern 5
        if(len(resultlist5)>0):
            count5 = count5 + len(resultlist5)
            for result in resultlist5:
                result_5.append(result[0])

#Displaying the phrases belonging to each pattern and its count

print("Phrases found for Hearst pattern 1 :\n",result_1)        
print("\nNumber of matching phrases for Hearst pattern 1: ", count1)

```
